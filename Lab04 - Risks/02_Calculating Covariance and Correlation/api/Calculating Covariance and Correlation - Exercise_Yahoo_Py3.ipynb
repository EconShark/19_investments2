## Calculating Covariance and Correlation

Consider a portfolio composed of *Walmart* and *Facebook*. Do you expect the returns of these companies to show high or low covariance? Or, could you guess what the correlation would be? Will it be closer to 0 or closer to 1? 

Begin by extracting data for Walmart and Facebook from the 1st of January 2014 until today.

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['WMT', 'META']

prices = yf.download(
    tickers,
    start="2014-01-01",
    auto_adjust=True
)['Close']

prices.head()
```

Repeat the process we went through in the lecture for these two stocks. How would you explain the difference between their means and their standard deviations?

```python
WMT_annual_return = prices["WMT"].pct_change().mean() * 252
META_annual_return = prices["META"].pct_change().mean() * 252
WMT_risk_d = np.log(prices["WMT"]).diff().std()
WMT_risk_a = WMT_risk_d * np.sqrt(252)
META_risk_d = np.log(prices["META"]).diff().std()
META_risk_a = META_risk_d * np.sqrt(252)

print(f"Walmart Annual return: {WMT_annual_return:.2%}")
print(f"Walmart Daily risk: {WMT_risk_d:.2%}")
print(f"Walmart Annual risk: {WMT_risk_a:.2%}\n")

print(f"Meta Annual return: {META_annual_return:.2%}")
print(f"Meta Daily risk: {META_risk_d:.2%}")
print(f"Meta Annual risk: {META_risk_a:.2%}")
```

```python
print("Meta offers higher average returns but at the cost of significantly higher volatility, while Walmart provides lower returns with more stable and predictable performance.")
```

## Covariance and Correlation


\begin{eqnarray*}
Covariance Matrix: \  \   
\Sigma = \begin{bmatrix}
        \sigma_{1}^2 \ \sigma_{12} \ \dots \ \sigma_{1I} \\
        \sigma_{21} \ \sigma_{2}^2 \ \dots \ \sigma_{2I} \\
        \vdots \ \vdots \ \ddots \ \vdots \\
        \sigma_{I1} \ \sigma_{I2} \ \dots \ \sigma_{I}^2
    \end{bmatrix}
\end{eqnarray*}

Covariance matrix:

```python
cov_matrix = prices.pct_change().cov()
print(f"Covariance Matrix:\n{cov_matrix}")
```

Correlation matrix:

```python
corr_matrix = prices.pct_change().corr()
print(f"Correlation Matrix:\n{corr_matrix}")
```
