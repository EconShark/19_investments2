## Calculating Portfolio Risk

Calculate the risk of an equally weighted portfolio composed of Microsoft and Apple. The data can be obtained from Yahoo Finance for the period 1st of January 2007 until today. 

*Hint: The code we went through in the lecture is what you need here. You will need to import the data first. The previous lessons could be a good reference point for that! :) *

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['MSFT', 'AAPL']

data = yf.download(
    tickers,
    start="2007-01-01",
    auto_adjust=True
)['Close']

data.head()
```

## MSFT

```python
MSFT_annual_return = data["MSFT"].pct_change().mean() * 252
MSFT_risk_d = np.log(data["MSFT"]).diff().std()
MSFT_risk_a = MSFT_risk_d * np.sqrt(252)

print(f"Microsoft Annual return: {MSFT_annual_return:.2%}")
print(f"Microsoft Daily risk: {MSFT_risk_d:.2%}")
print(f"Microsoft Annual risk: {MSFT_risk_a:.2%}")
```

## Apple

```python
AAPL_annual_return = data["AAPL"].pct_change().mean() * 252
AAPL_risk_d = np.log(data["AAPL"]).diff().std()
AAPL_risk_a = AAPL_risk_d * np.sqrt(252)

print(f"Apple Annual return: {AAPL_annual_return:.2%}")
print(f"Apple Daily risk: {AAPL_risk_d:.2%}")
print(f"Apple Annual risk: {AAPL_risk_a:.2%}")
```

## Covariance and Correlation

\[
\begin{aligned}
\text{Covariance Matrix: } \quad
\Sigma &=
\begin{bmatrix}
\sigma_1^2 & \sigma_{12} & \dots & \sigma_{1I} \\
\sigma_{21} & \sigma_2^2 & \dots & \sigma_{2I} \\
\vdots & \vdots & \ddots & \vdots \\
\sigma_{I1} & \sigma_{I2} & \dots & \sigma_I^2
\end{bmatrix}
\end{aligned}
\]

*****

Covariance matrix:

```python
cov_matrix = np.log(data).diff().cov()
print(f"Covariance Matrix:\n{cov_matrix}")
```

***

Correlation matrix:

```python
corr_matrix = data.pct_change().corr()
print(f"Correlation Matrix:\n{corr_matrix}")
```

## Calculating Portfolio Risk

Equal weigthing scheme:

```python
weights = np.array([0.5, 0.5])
```

Portfolio Variance:

```python
portfolio_variance = np.dot(weights.T, np.dot(cov_matrix*252, weights))
print(f"Portfolio Variance: {portfolio_variance:.2%}")
```

Portfolio Volatility:

```python
porfolio_volatility = np.sqrt(portfolio_variance)
```

```python
print(f"Portfolio Volatility: {porfolio_volatility:.2%}")
```
