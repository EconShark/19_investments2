## Calculating the Risk of a Security

*Suggested Answers follow (usually there are multiple ways to solve a problem in Python).*

Download the data for Microsoft (‘MSFT’) from Yahoo Finance for the period ‘2000-1-1’ until today. <br />
Assess the daily and the annual risk of ‘MSFT’. Repeat the exercise for Apple for the same timeframe. 

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['MSFT', 'AAPL']

prices = yf.download(
    tickers,
    start="2000-01-01",
    auto_adjust=True
)['Close']

prices.head()
```

### MSFT

Daily risk:

```python
MSFT_risk_d = np.log(prices["MSFT"]).diff().std()
MSFT_risk_d
```

Annual risk:

```python
MSFT_risk_a = MSFT_risk_d * np.sqrt(252)
MSFT_risk_a
```

### Apple

Daily risk:

```python
AAPL_risk_d = np.log(prices["AAPL"]).diff().std()
AAPL_risk_d
```

Annual risk:

```python
AAPL_risk_a = AAPL_risk_d * np.sqrt(252)
AAPL_risk_a
```

******

Store the volatilities of the two stocks in an array called “vols”.

```python
vols=np.array([MSFT_risk_a, AAPL_risk_a])
vols
```

How are the two risk values different?

AAPL has higher return variability than MSFT, so it carries higher statistical risk in your dataset.
