## Logarithmic Returns

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

MSFT = yf.download(
    "MSFT",
    start="2000-01-01",
    auto_adjust=True
)

MSFT["Simple Return"] = MSFT["Close"].pct_change()

print (MSFT['Simple Return'])
```

### Log Returns

$$
ln(\frac{P_t}{P_{t-1}})
$$

Calculate the Log returns for Microsoft.

```python
MSFT["Log Return"] = np.log(MSFT["Close"]).diff()
print(MSFT["Log Return"])
```

Plot the results on a graph.

```python
MSFT["Log Return"].plot(figsize=(10, 6))
```

Estimate the daily and the annual mean of the obtained log returns.

```python
average_log_return_d = MSFT["Log Return"].mean()
```

```python
average_log_return_a = average_log_return_d*252
```

Print the result in a presentable form.

```python
print(f"Average Log Return (d): {average_log_return_d * 100:.2f} %")
print(f"Average Log Return (a): {average_log_return_a * 100:.2f} %")
```
