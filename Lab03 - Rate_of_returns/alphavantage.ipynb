```python
import requests
import pandas as pd

symbols = ["IBM", "TSCO.LON", "SHOP.TRT"]
api_key = "demo"

series_list = []

for symbol in symbols:

    params = {
        "function": "TIME_SERIES_DAILY_ADJUSTED",
        "symbol": symbol,
        "outputsize": "full",
        "apikey": api_key
    }

    r = requests.get(
        "https://www.alphavantage.co/query",
        params=params
    )

    data = r.json()

    df = pd.DataFrame(data["Time Series (Daily)"]).T
    df.index = pd.to_datetime(df.index)
    df = df.sort_index()

    series_list.append(df["5. adjusted close"])

pfolio_av = pd.concat(series_list, axis=1)
pfolio_av.columns = symbols

print(pfolio_av.head())
```
