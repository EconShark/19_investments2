## Simple Returns - Part II

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

MSFT = yf.download(
    "MSFT",
    start="2000-01-01",
    auto_adjust=True
)

MSFT["Simple Return"] = MSFT["Close"].pct_change()

print (MSFT['Simple Return'])
```

Plot the simple returns on a graph.

```python
MSFT["Simple Return"].plot(figsize=(10, 6))
plt.title("Simple Return of MSFT")
```

Calculate the average daily return.

```python
average_return_d = MSFT["Simple Return"].mean()
print(f"Average Simple Return: {average_return_d:.4f}")
```

Estimate the average annual return.

```python
average_return_a = MSFT["Simple Return"].mean() * 252
print(f"Average Annualized Simple Return: {average_return_a:.4f}")
```

Print the percentage version of the result as a float with 2 digits after the decimal point.

```python
print(f"Average of annualized simple return as a float with 2 digits after the decimal point.: {average_return_a*100:.2f} %")
```
